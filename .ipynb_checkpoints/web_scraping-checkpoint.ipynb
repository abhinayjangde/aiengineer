```python
import requests
import pandas as pd
```

```python
url = "https://jsonplaceholder.typicode.com/todos"
res = requests.get(url)
```

```python
json_data = res.json()
df = pd.json_normalize(json_data)

df = df[['id', 'title', 'completed']]
```

```python
df = df[df['completed'] == True]
```

```python
URL = "https://www.scrapethissite.com/pages/simple"
res = requests.get(URL)
```

```python
if res.status_code == 200 :
    print(res.headers)

with open("scraped_data/data1.html", "w") as f:
    f.write(res.text)

```

```python

for i in range(1,8):
    res = requests.get(f"https://quotes.toscrape.com/page/{i}")
    if res.status_code == 200:
        with open(f"scraped_data/quote{i}.html", "w") as f:
            f.write(res.text)
```
