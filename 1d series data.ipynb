```python
age = pd.Series([33,24,25], index=["Adam", "Bob", "Charlie"])
```

```python
print(age)
```

```python
print(age["Bob"])
print(age.iloc[1]) # access by location
```

```python
# series are homogenous
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([10,20,30,40])
```

```python
print(s1 + s2)
```

```python
# pandas series are mutable values
print(s1)
s1[0] = 99
print(s1)
```

```python
# drop the specific index value using drop() function and it return new series

changed = s1.drop(0)
print(changed)
```
