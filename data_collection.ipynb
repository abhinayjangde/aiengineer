```python
# data collection throught APIs
import requests
```

```python
url = "https://stephen-king-api.onrender.com/api/books"

res = requests.get(url)

json_data = res.json()
```

```python
import pandas as pd

df = pd.json_normalize(json_data['data'])
```

```python
df = df[['id', 'Title', 'Year', 'ISBN']]
```
