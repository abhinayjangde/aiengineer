```python
import numpy as np
import time
```

```python
size = 100_000_00

list_arr = list(range(size))

start = time.time()
sqrt_list = [x ** 2 for x in list_arr]
end = time.time()
print(f"list time {end - start } seconds")

np_arr = np.array(list_arr)

start = time.time()
new_arr = np_arr ** 2
end = time.time()
print(f"list time {end - start } seconds")
```
